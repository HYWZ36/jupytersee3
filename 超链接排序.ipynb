```python
import os
import os.path

txt_path = 'C:/Users/wz/Desktop/quchong1/'
des_txt_path = 'C:/Users/wz/Desktop/quchong2/'

txt_files = os.listdir(txt_path)  # txt_pathtxttxt


def sort_txt_content():
    for txtfile in txt_files:
        if not os.path.isdir(txtfile):
            out_file = open(des_txt_path + txtfile, 'a')  # 'a''\n'
            in_file = open(txt_path + txtfile, 'r')
            lines = in_file.readlines()
            b = [[line.split(" ")[0]] for line in lines]  # b
            lines.sort(key=lambda b: len(b))

            for line in lines:
                print(line)
                out_file.write(line)  # txt


if __name__ == '__main__':
    sort_txt_content()
```
